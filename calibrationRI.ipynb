```python
Me_ratios = [0.1, 0.3, 0.5, 0.7, 0.9]
RI = [1.374, 1.369, 1.362, 1.3495, 1.341]
```

```python
import matplotlib.pyplot as plt

plt.plot(Me_ratios, RI)
plt.xlabel('Me mol ratio') # label on the x axis
plt.ylabel('RI') # label on the y axis
plt.title('Experimental RI values')
plt.show()
```

# Calibration => liner regression model 

```python
import numpy as np
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(Me_ratios, RI)

def func_linregr(x):
  return slope * x + intercept

calibration = list(map(func_linregr, Me_ratios))

plt.scatter(Me_ratios, RI, label='experimental data')
plt.plot(Me_ratios, calibration, 'r')
plt.xlabel('Methanol mol fractions')
plt.ylabel('RI')
plt.title('Calibration Curve')
plt.legend()
plt.show()
```
