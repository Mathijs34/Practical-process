```python
from matplotlib.animation import FuncAnimation
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

def Cp_v_Me(T):
    Cp_v_Me =  4.853229866724107e-05*T**2 + 0.026393190719809542*T + 32.831657334235835
    return Cp_v_Me
def Cp_v_Iso(T):
    Cp_v_Iso = -6.900678097360619e-05*T**2 + 0.250658478269019*T + 21.727358906262694
    return Cp_v_Iso
def Cp_mix_y(Me_frac,T):
    Cp_mix_y = Me_frac*Cp_v_Me(T) + (1-Me_frac)*Cp_v_Iso(T)
    return Cp_mix_y

fig = plt.figure()

lines = plt.plot([])
line = lines[0]
x  = np.linspace(0, 1000, 1000)
# other setup 
plt.xlim(0,1000)
plt.ylim(0,200)
plt_1= plt.plot(x,Cp_v_Me(x))
plt_2= plt.plot(x,Cp_v_Iso(x))
plt.ylabel("Heat capacity (J/(mol*K)")
plt.xlabel("Temperature (K)")
plt.title("Cp Gas Methanol-Isopropanol Mixture")

def animate(frame):
    # update plot
    y = Cp_mix_y((np.abs((frame/240)*2-1)),x)
    line.set_data((x, y))

plt.legend(["Cp mix", "Cp MeOH", "Cp Iso"])
anim = FuncAnimation(fig, animate, frames=240, interval=20)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
