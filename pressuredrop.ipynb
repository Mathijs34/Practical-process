```python
# Pressure drop batch
import numpy as np
import matplotlib.pyplot as plt
```

```python
power = np.linspace(250,1500,6) #W
boilup_rate=[0.0394, 0.2474, 0.4941, 0.7561, 0.9872, 1.188] #mL/s
pressure_drop = [312.98, 518.37, 547.71, 567.27, 586.83, 606.39] #Pa

ln_br = np.log(boilup_rate)
ln_pd = np.log(pressure_drop)
```

```python
b= plt.plot(power, boilup_rate, 'o-')
plt.xlabel('Power (W)')
plt.ylabel('Boilup rate (mL/s)')
plt.grid(b)
plt.title('Boilup rate vs Power')
plt.show()
```

```python
c= plt.plot(ln_br, ln_pd, 'o-')
plt.xlabel('ln of boilup rate')
plt.ylabel('ln of pressure drop')
plt.grid(c)
plt.title('Pressure drop vs Boilup rate (logarithmic)')
plt.show()
```
